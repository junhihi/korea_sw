```python
URLs = list()
Seens = list()

URLs.append([{'url':'https://www.scrapingcourse.com/pagination',
              'depth':1}])

while URLs:
    seed = URLs.pop(-1) # queue
    Seens.append(seed)

    print(seed)

    # robots.txt 확인

    resp = request('get', seed['url'])

    if resp.status_code != 200:
        continue

    if not re.search(r'text/html', resp.headers['content-type']):
        continue
    
    # 각 페이지의 이미지도 함께 수집
    # headers['content-type']
    # dom.select('a')
    # Focused: 전략

    dom = BeautifulSoup(resp.text, 'html.parser')

    for a in dom.select('a[href], .product-image'):
        link = a.attrs['href'] if a.has_attr('href') else a.attrs['src']
        if not re.match(r'#|javascript', link):
            nurl = urljoin(seed['url'], link)
            # Focused: 도메인
            if nurl not in [s['url'] for s in Seens] and\
            nurl not in [s['url'] for s in URLs]: 
                URLs.append({'url':nurl, 'depth':seed['depth']+1}) 
```
