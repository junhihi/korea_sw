```python
HTTP - urllib - requests
Client ---> HTTP ---> Server
    Request     Response    => TCP/ip
     Header     Body        => Bytes

Opt-out(Blacklist -> 거부를 명시적으로 표현) => 기록을 남겨야 함
1. Robots.txt => Bot (agent)가 Enable, Disable 명시적으로 표현
   ; urlib.robotparser 잘 안돌아감
   ; HTML-HEAD-META noindex, nofollow 확인
2. 트래픽 => 과도한 요청 X => delay
3. 개인정보 => 그 어떤 경우에도 X
4. 저작권 => DB에 있는 자료 X
5. 약관 => 현실적으로 어려움 (약관을 익힌 ai)
6. 안전한 방법 => API
공부용, 연구용, 내 데이터 내가 수기, 공유 X
```

```python
urlopen(url=str or Request)
GET, 파라미터를 전달하는 방법 => URL ? key=value & ... (Bytes)
Percent Encoding(Hexadecimal)
=> quote, unquote

HTML내에서 표현하는 방법 => HTML Entities => &[value];, &10진수;
=> escape, unescape

HTTP.Response 객체로 돌려줌
Status Code, Status Reason, Headers[content-type]
200=OK, 400(Client Error) => 멈춤, 500(Server Error) => 다시 한번 시도 OK

?q=카리나
            Naver       Daum        Google
robots.txt    X           X            X
실제 해보니   아무 문제     captcha     javascript
user-agent  아무 문제    아무 문제      javascript => cookie

?q=카리나 => quote('한글') => %AD...
?k1=v1&k2=v2&k3=v3... => params 관리, URL 분리/합치
```

```python
from urllib.error import HTTPError

url = 'https://httpbin.org/status/500'
try:
    urlopen(url)
except HTTPError as e:
    # if e.status != 200:
    if 400 <= e.status < 500:
        print(e.status)
        print(e.reason)
        print(e.headers)
    elif 500 <= e.status:
        print('try again')
```
